## Assignment 3
Consider an infinite number of coins, each with a value from a set $M \subset \mathbb{N}$. A given value $B \in \mathbb{N}$ should now be represented by as few coins as possible. For this the following algorithm is used: Choose the largest coin value that is smaller than or equal to $B$ and subtract it from $B$. This is repeated until the value is $0$. The algorithm then outputs the required coins.

(a) Prove the following statement by induction: The algorithm always provides an optimal solution with $M = \{1, 5, 10\}$ and $B \in \mathbb{N}$.

**Hint:** Consider the 3 cases $B < 5$, $5 \leq B < 10$ and $10 \leq B$.

*Proof.* Two obervations:
1. An optimal superset must not have $> 4$ 1-coins or $> 1$ 5-coins.
2. For any B, there is only one possible multiset with $< 5$ 1-coins and $< 2$ 5-coins.

$\implies$ A multiset with $< 5$ 1-coins and $< 2$ 5-coins is *optimal*.  
The greedy algorithm does not take $> 4$ 1-coins or $> 1$ 5-coins. Therefore, the greedy algorithm is optimal.

(b) Prove or disprove the following statement: The algorithm always provides an optimal solution with any 3 coin values as long as one of the coin values is $1$.

Claim: The algorithm does **not** yield an optimal solution with these constraints.


*Proof.* Let $M = \{1, 3, 4\}$ and $B = 6$. The descibed greedy algotithm would choose 
1. Use 4, update $B = 6 - 4 = 2$
2. Use 1, update $B = 2 - 1 = 1$
3. Use 1, update $B = 1 - 1 = 0$


This procedure takes **3 coins**. But $6 = 3 + 3$ does only need **2 coins**.

(c) Implement this algorithm.

```python
def greedy_coin_change(M, B):
    coins = []
    while B > 0:
        largest = max(c for c in M if c <= B)
        coins.append(largest)
        B -= largest
    return coins
```

```python
test(greedy_coin_change([10, 5, 1], 0), [])
test(greedy_coin_change([1, 5, 10], 8), [5, 1, 1, 1])
test(greedy_coin_change([5, 1, 10], 27), [10, 10, 5, 1, 1])
```
