$$
n! > 2^n ~~\forall n \in \mathbb{N}, n \geq 4
$$

Induction start, $n = 4$:
$$4! = 24 > 16 = 2^4$$
Induction step, $n \to n + 1$:
$$
\begin{aligned}
(n + 1)! &= (n + 1) \cdot n! \\
&\geq 2 \cdot n!  && \text{since $n \geq 4$} \\
&\geq 2 \cdot 2^n && \text{assumtion of the induction step} \\
&= 2^{n+1}
\end{aligned}
$$

## Assignment 3

Given in the following is a text $T = t_0 t_1 t_2 \dots t_{n-1}$, a pattern $s = s_0 s_1 s_2 \dots s_{m-1}$, $m \leq n$, and a number $k \in \mathbb{N}$.<br><br>
The Hamming distance $d_H(s, s')$ between two strings is defined as the number of positions $i$ at which $s_i \neq s_i'$.

1. Implement an algorithm that calculates the Hamming distance $d_H(s, s')$ between two strings $s$ and $s'$. You may assume that $|s| = |s'|$, where $|x|$ is the length of a string $x$.

```python
def dH(v, w):
    if len(v) != len(w):
        raise ValueError("Strings must be of same length!")
    
    return sum(a != b for a,b in zip(v, w))
```

```python
test(dH('', ''), 0)
test(dH('abc', 'abz'), 1)
test(dH('abc', 'cba'), 2)
test(dH('aabc', 'abca'), 3)
```

2. Implement an algorithm that finds the index of the first occurrence of a string $s'$ in $T$ such that $d_H(s, s') \leq k$. If there is no such $s'$, $-1$ should be returned.

```python
def pattern_matching_with_k_errors(text, pattern, k):
    # Get empty pattern special case out of the way
    if not pattern:
        return 0
    pattern_length = len(pattern)
    for i in range(len(text) - pattern_length + 1):
        if (dH(text[i:i + pattern_length], pattern) <= k):
            return i
    return -1
```

```python
test(pattern_matching_with_k_errors('AGCGTATTCGGTAGC', 'AGGTG', 2), 8)
test(pattern_matching_with_k_errors('', 'A', 0), -1)
test(pattern_matching_with_k_errors('AGCGTATTCGGTAGC', 'TAGC', 0), 11)
test(pattern_matching_with_k_errors('AGCGTATTCGGTAGC', 'TAGC', 2), 4)
```

Let $\Sigma = \{\mathit{A, C, G, T}\}$ be an alphabet. A $k$-mer is an element $s \in \Sigma^k$ (a string of length $k$).

Let $x$ be a $k$-mer. We define the $d$-neighbourhood of x as the set of all $k$-mers that have a Hamming distance of at most $d$ to $x$.

For example, for the $3$-mer $\mathit{ACG}$ and $d=1$ the $d$-neighbourhood is: $\{\mathit{AAG, ACA, ACC, ACG, ACT, AGG, ATG, CCG, GCG, TCG}\}$.

Implement an algorithm that calculates the $d$-neighbourhood for a $k$-mer $x$ and a non-negative integer $d$.

```python
def dNeighborhood(x, d):
    if d < 0:
        raise ValueError("Distance d must be non negative")
    
    if not x:
        return {""}
    sigma = ["A", "C", "G", "T"]
    
    neighbors = set()
    for replacement in sigma:
        changed = replacement != x[0]
        if (d == 0 and changed):
            continue
        for suffix in dNeighborhood(x[1:], d-1 if changed else d):
            neighbors.add(replacement + suffix)

    return neighbors
```

```python
test(set(dNeighborhood('ACG', 1)), set(['AAG', 'ACA', 'ACC', 'ACG', 'ACT', 'AGG', 'ATG', 'CCG', 'GCG', 'TCG']))
test(set(dNeighborhood('TTT', 1)), set(['ATT', 'CTT', 'GTT', 'TAT', 'TCT', 'TGT', 'TTA', 'TTC', 'TTG', 'TTT']))
test(set(dNeighborhood('TTT', 2)), set(['AAT', 'ACT', 'AGT', 'ATT', 'CAT', 'CCT', 'CGT', 'CTT', 'GAT', 'GCT', 'GGT', 'GTT', 'TAT', 'TCT', 'TGT', 'TTT', 'TAA', 'TAC', 'TAG', 'TAT', 'TCA', 'TCC', 'TCG', 'TCT', 'TGA', 'TGC', 'TGG', 'TGT', 'TTA', 'TTC', 'TTG', 'ATA', 'ATC', 'ATG', 'ATT', 'CTA', 'CTC', 'CTG', 'CTT', 'GTA', 'GTC', 'GTG', 'GTT']))
```
