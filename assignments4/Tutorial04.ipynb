## Assignment 1

Implement a function that outputs the $n$-th Fibonacci number. It should be based on dynamic programming.<br>
The Fibonacci numbers are defined as follows:
$f_0 = 0, f_1 = 1,$ and $f_n = f_{n-1} + f_{n-2}$ for $n \geq 2$.

```python
def fibonacci(n: int):
    last = 1
    second_last = 0
    for _ in range(n):
        new_last = last + second_last
        second_last = last
        last = new_last
    return second_last
```

```python
test(fibonacci(6), 8)
test(fibonacci(7), 13)
```

## Assignment 2

The Manhattan-Tourist problem was introduced on graphs corresponding to a rectangular grid of size $n \times m$.

(a) Implement a function that solves the Manhattan-Tourist problem and is based on dynamic programming.<br>
The input is a grid graph in the form of a dictionary that maps a tuple of nodes to a weight, as well as the sizes $n$ and $m$ of the graph (see next cell and the test cases). For this graph, the length of the longest path (i.e. the sum of the edge weights on the path) is to be output from the start field (top left) to the end field (bottom right).

```python
from collections import defaultdict
g1 = defaultdict(lambda: 1)
g1[((0, 0), (0, 1))] = 0
g1[((0, 2), (0, 3))] = -3
g1[((0, 4), (1, 4))] = 9
g1[((1, 1), (2, 1))] = 5
g1[((1, 4), (2, 4))] = 5
g1[((2, 1), (2, 2))] = 2
g1[((2, 2), (2, 3))] = 4
g1[((3, 1), (4, 1))] = 5
g1[((4, 2), (4, 3))] = 3
g2 = defaultdict(lambda: 1)
g2[((0, 0), (0, 1))] = 3
g2[((1, 1), (1, 2))] = 5
g2[((1, 4), (1, 5))] = 3
g2[((2, 5), (2, 6))] = 3
g2[((2, 0), (3, 0))] = 2
g2[((3, 3), (3, 4))] = 2
g2[((3, 5), (4, 5))] = 7
g2[((4, 5), (4, 6))] = 2
g2[((4, 6), (4, 7))] = 4
g2[((4, 7), (4, 8))] = 5
g2[((5, 1), (5, 2))] = 3
g2[((5, 5), (6, 5))] = 8
g2[((6, 5), (6, 6))] = 2
g2[((6, 2), (7, 2))] = 10
g2[((6, 6), (7, 6))] = -2
g2[((7, 7), (7, 8))] = 3
```

```python
def solve_mtp_dp(g, n, m):
    dp = [[0] * (m+1) for _ in range(n+1)]
    for i in range(n+1):
        for j in range(m+1):
            if i == 0 and j == 0:
                dp[i][j] = 0
                continue
            if j == 0:
                dp[i][j] = dp[i-1][j] + g[((i-1, j), (i, j))]
                continue
            elif i == 0:
                dp[i][j] = dp[i][j-1] + g[((i, j-1), (i, j))]
                continue
            dp[i][j] = max(
                dp[i-1][j] + g[((i-1, j), (i, j))], 
                dp[i][j-1] + g[((i, j-1), (i, j))]
                )
    return dp[n][m]
```

```python
test(solve_mtp_dp(g1, 4, 4), 18)
test(solve_mtp_dp(g2, 8, 8), 40)
```

(c) Implement a function that outputs all paths from part (b). Input are the parameters $n$ and $m$, output should be a list of paths, where each path should be a list of coordinates corresponding to the nodes. The first coordinate is always the start node $(0, 0)$ and the last is always the destination node $(n, m)$. The order in which the paths are output is arbitrary.

```python
def enumerate_paths(n, m):
    all_paths = []
    def partials_paths(path, i, j):
        if i == 0 and j == 0:
            all_paths.append(path[::-1])   # Test assumes ascending order
            return
        if i == 0:
            path.append((i, j - 1))
            partials_paths(path, i, j - 1)
            return
        if j == 0:
            path.append((i - 1, j))
            partials_paths(path, i - 1, j)
            return
        path1 = path.copy()
        path2 = path.copy()
        path1.append((i - 1, j))
        path2.append((i, j - 1))
        partials_paths(path1, i - 1, j)
        partials_paths(path2, i, j - 1)
    partials_paths([(n, m)], n, m)
    return all_paths
```

```python
test(sorted(enumerate_paths(1, 1)), [[(0, 0), (0, 1), (1, 1)], [(0, 0), (1, 0), (1, 1)]])
test(sorted(enumerate_paths(1, 2)), [[(0, 0), (0, 1), (0, 2), (1, 2)], [(0, 0), (0, 1), (1, 1), (1, 2)], [(0, 0), (1, 0), (1, 1), (1, 2)]])
test(sorted(enumerate_paths(0, 3)), [[(0, 0), (0, 1), (0, 2), (0, 3)]])
test(sorted(enumerate_paths(0, 0)), [[(0, 0)]])
test(len(enumerate_paths(6, 3)), 84)
test(len(enumerate_paths(3, 6)), 84)
```

(d) Implement the function below that solves the Manhattan-Tourist problem.<br>
Input and output are as in (a). For the input graph, all possible paths should be generated and the length of the longest path (i.e. the sum of the edge weights on the path) should be output.

```python
def solve_mtp(g, n, m):
    all_paths = enumerate_paths(n, m)
    best_score = float('-inf')
    for path in all_paths:
        score = 0
        for step in range(len(path) - 1):
            score += g[(path[step], path[step + 1])]
        best_score = max(best_score, score)
    return best_score
```

```python
test(solve_mtp(g1, 4, 4), 18)
test(solve_mtp(g2, 8, 8), 40)
```

## Assignment 3

In this exercise we revisit the coin-change problem from Tutorial 2 Assignment 3. Given an amount of money $M \in \mathbb{N}$, find the smallest number of coins of denominations $c_1,\dots,c_n \in \mathbb{N}$ into which you can change $M$.<br>
As we have seen, the greedy algorithm is not guaranteed to yield the optimal solution.
This time we want to approach the problem using dynamic programming.

(a) Formulate the recurrence relation for the problem of finding the **smallest number of coins**  into which you can change $M$.

Take the min of
- $(M - C_1) + 1$
- $(M - C_2) + 1$
- ...
- $(M - C_n) + 1$

(b) Implement a function that solves the coin-change problem using your results from (a).

```python
def dp_coin_change(coins: list[int], M:int) -> int:
    min_dp = {0: 0}
    def dp_coin_change_rec(coins: list[int], M:int) -> int:
        if M == 0:
            return 0
        if M in min_dp:
            return min_dp[M]

        m = float('inf')
        for coin in coins:
            if M - coin >= 0:
                m = min(m, dp_coin_change_rec(coins, M - coin))
        m += 1
        min_dp[M] = m
        return m
    return dp_coin_change_rec(coins, M)
```

```python
test(dp_coin_change([1, 5, 10], 8), 4)
test(dp_coin_change([10, 5, 1], 0), 0)
test(dp_coin_change([1, 3, 4], 6), 2)
test(dp_coin_change([1, 11, 90], 99), 9)
```

(c) What is the asymptotic runtime of your algorithm? Give reasons for your answer.

O(M) - at max M entries are computed

(d) Adapt your function from (b) to output the coin values that make up an optimal solution.

> Hint: You may want to use backtracking similar to how the longest path was constructed in the Manhatten Tourist Problem.

```python
def adapted_dp_coin_change(coins: list[int], M:int) -> int:
    min_dp = {0: 0}
    min_direction = {0: None}
    
    def dp_coin_change_rec(coins: list[int], M:int) -> int:
        if M == 0:
            return 0
        if M in min_dp:
            return min_dp[M]

        m = float('inf')
        best_coin = None
        for coin in coins:
            if M - coin < 0:
                continue
            option = dp_coin_change_rec(coins, M - coin)
            if option < m:
                m = option
                best_coin = coin
        m += 1
        min_dp[M] = m
        min_direction[M] = best_coin
        return m
    
    dp_coin_change_rec(coins, M)

    rest = M
    change = []
    while rest > 0:
        coin = min_direction[rest]
        change.append(coin)
        rest -= coin
    return change
```

```python
test(sorted(adapted_dp_coin_change([10, 5, 1], 0)), [])
test(sorted(adapted_dp_coin_change([1, 5, 10], 8)), [1, 1, 1 , 5])
test(sorted(adapted_dp_coin_change([1, 3, 4], 6)), [3, 3])
test(sorted(adapted_dp_coin_change([1, 11, 90], 99)), [11]*9)
```
